```python
import json
import jieba
```

```python
#词频统计
totalWordFreq = {}
for num in range(341):
    a = json.load(open('155/0/'+str(num)+'.json','r'))
    for i in a:
        for word in i:
            if word in totalWordFreq:
                totalWordFreq[word] += 1
            else:
                totalWordFreq[word] = 1

for num in range(70):
    a = json.load(open('165/0/'+str(num)+'.json','r'))
    for i in a:
        for word in i:
            if word in totalWordFreq:
                totalWordFreq[word] += 1
            else:
                totalWordFreq[word] = 1

for num in range(168):
    a = json.load(open('166/0/'+str(num)+'.json','r'))
    for i in a:
        for word in i:
            if word in totalWordFreq:
                totalWordFreq[word] += 1
            else:
                totalWordFreq[word] = 1

for num in range(183):
    a = json.load(open('167/0/'+str(num)+'.json','r'))
    for i in a:
        for word in i:
            if word in totalWordFreq:
                totalWordFreq[word] += 1
            else:
                totalWordFreq[word] = 1

for num in range(171):
    a = json.load(open('168/0/'+str(num)+'.json','r'))
    for i in a:
        for word in i:
            if word in totalWordFreq:
                totalWordFreq[word] += 1
            else:
                totalWordFreq[word] = 1

for num in range(53):
    a = json.load(open('169/0/'+str(num)+'.json','r'))
    for i in a:
        for word in i:
            if word in totalWordFreq:
                totalWordFreq[word] += 1
            else:
                totalWordFreq[word] = 1
```

```python
freq_word = []
for word, freq in totalWordFreq.items():
        freq_word.append((word, freq))
freq_word.sort(key = lambda x: x[1], reverse = True)
```
