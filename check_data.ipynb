```python
import json
import pandas as pd
from os import path
```

```python
label_fn = "label/155label.csv"
# label_fn = "label/155label_0.csv"
# label_fn = "label/156label.csv"
label = pd.read_csv(label_fn, index_col=0, header=0)

label
```

```python
for i in label.index:
    l = label.loc[i]
    # if pdf has no label, ignore this pdf
    if l.count() == 0:
        continue
    fn = "2015-5/json5/" + "15" + str(i) + ".json"
    # if 2016-5/5json, use 
    #fn = "2016-5/5json/" + "5" + str(i-1) + ".json"
    if not path.exists(fn):
        print(i, "json file not exists!")
        continue
    f = open(fn, "r").read()
    c = json.loads(f)
    if l.count() != len(c):
        print("in label csv file line json_num: %d" % i, "label length does not match pdf length")
```
